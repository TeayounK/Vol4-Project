```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
```

```python
tau = .085
beta = 0.7
gamma = 0.36
alpha = 0.75
delta = .25

# Make the ODE function
def ode(t, y):
    '''defines the ode system'''
    return np.array([tau*y[2] + tau*y[1] - beta*y[0]-delta*y[0],
                     alpha*y[2] + beta*y[0] - gamma*y[1] - tau*y[1],
                     gamma*y[1] - alpha*y[2] - tau*y[2] + delta*y[0]])
# Make the time domain
t_span = (0, 10)
# Set the initial conditions
y0 = np.array([1,0,0])
# solve the system
sol = solve_ivp(ode, t_span, y0, max_step=0.05)

# Plot solution and compare to actual
plt.subplot(111)
plt.plot(sol.t,sol.y[0], label="G")
plt.plot(sol.t,sol.y[1], label="B")
plt.plot(sol.t,sol.y[2], label="I")
plt.legend()
plt.ylabel("Percent of total capital")
plt.xlabel("Time (years)")
plt.title("New Govenment, New Money")

plt.savefig("new_govt_money.png", dpi = 300)
```
