```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
tau = .085
beta = 0.7
gamma = 0.36
alpha = 0.75
delta = .25

# Make the ODE function
def ode(t, y):
    '''defines the ode system'''
    return np.array([tau*(gamma*y[1] + delta*y[0]) + tau*(alpha*y[2] + beta*y[0]) - beta*y[0]-delta*y[0],
                     alpha*y[2] + beta*y[0] - gamma*y[1] - tau*(alpha*y[2] + beta*y[0]),
                     gamma*y[1] - alpha*y[2] - tau*(gamma*y[1] + delta*y[0]) + delta*y[0]])
# Make the time domain
t_span = (0, 10)
# Set the initial conditions
y0 = np.array([1,0,0])
# solve the system
sol = solve_ivp(ode, t_span, y0, max_step=0.05)

# Plot solution and compare to actual
plt.subplot(111)
plt.plot(sol.t,sol.y[0], label="G")
plt.plot(sol.t,sol.y[1], label="B")
plt.plot(sol.t,sol.y[2], label="I")
plt.legend()
plt.ylabel("Percent of total capital")
plt.xlabel("Time (years)")
plt.title("New Govenment, New Money")

plt.savefig("new_govt_money.png", dpi = 300)
```

```python
tau_i = .085
tau_b = .5
beta = 0.7
gamma = 0.5
alpha = 0.75
delta = .25
i_i = 0.2
i_g = 0.5
# Make the ODE function
def ode(t, y):
    '''defines the ode system'''
    G = y[0]
    B = y[1]
    I = y[2]
    k1 = i_g/B
    k3 = i_i/B
    # Solve explicit parts of the ODE
    v1 = tau_i*(gamma*B + delta*G) + tau_b*(alpha*I + beta*G) - beta*G - delta*G - i_g*G
    v2 = alpha*I + beta*G - gamma*B - tau_b*(alpha*I + beta*G) + i_g*G + i_i*I
    v3 = gamma*B + delta*G - alpha*I - tau_i*(gamma*B + delta*G) - i_i*I
    # Find cubic coefficients
    a = k1*k3
    b = k1 + k3 + k1*k3*(I + G - v1 - v2 - v3)
    c = 1 + k1*G + k3*I - k1*v1 - k3*v3 - (k1 + k3)*v2
    d = -v2
    # Solve cubic for B_dot
    p = -b/(3*a)
    q = p**3 + (b*c - 3*a*d)/(6*a**2)
    r = c/(3*a)
    B_dot = (q + np.emath.sqrt(q**2 + (r-p**2)**3))**(1./3) + (q - np.emath.sqrt(q**2 + (r-p**2)**3))**(1./3) + p
    # Get G_dot and I_dot from B_dot value
    G_dot = (v1+k1*G*B_dot)/(1+k1*B_dot)
    I_dot = (v3+k3*I*B_dot)/(1+k3*B_dot)
    return np.array([G_dot, B_dot, I_dot])
def old_ode(t, y):
    '''defines the ode system'''
    return np.array([tau_i*(gamma*y[1] + delta*y[0]) + tau_b*(alpha*y[2] + beta*y[0]) - beta*y[0]-delta*y[0],
                     alpha*y[2] + beta*y[0] - gamma*y[1] - tau_b*(alpha*y[2] + beta*y[0]),
                     gamma*y[1] - alpha*y[2] - tau_i*(gamma*y[1] + delta*y[0]) + delta*y[0]])
# Make the time domain
t_span = (0, 10)
# Set the initial conditions
y0 = np.array([10.,28.,90.])
y0_investing = np.array([10.*(1-i_g),28.+10*i_g+90*i_i,90.*(1-i_i)])
y0 /= la.norm(y0,1)
y0_investing /= la.norm(y0_investing,1)
# solve the system
sol = solve_ivp(ode, t_span, y0, max_step=0.05)
sol2 = solve_ivp(old_ode, t_span, y0, max_step=0.05)

# Plot solution and compare to actual
plt.subplot(111)
plt.plot(sol.t,sol.y[0]*(1+i_g), "b-", label="G")
plt.plot(sol.t,sol.y[1]*1-sol.y[0]*i_g-sol.y[2]*i_i, "g-", label="B")
plt.plot(sol.t,sol.y[2]*(1+i_i), "r-", label="I")
plt.plot(sol.t,sol2.y[0], "b--", label="G (w/o investing)")
plt.plot(sol.t,sol2.y[1], "g--", label="B (w/o investing)")
plt.plot(sol.t,sol2.y[2], "r--", label="I (w/o investing)")
plt.legend()
plt.ylabel("Percent of total capital")
plt.xlabel("Time (years)")
plt.title("Starting from Current US Values, 50% Business Tax")

plt.savefig("investing_1.png", dpi = 300)
```
